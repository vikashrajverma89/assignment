# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a Python data visualization library based on Matplotlib that provides a high-level interface for creating informative and attractive statistical graphics. Here are five commonly used plots in Seaborn along with their typical use cases:

1. **Scatter Plot (`sns.scatterplot`):**
   - **Use:** Scatter plots are used to visualize the relationship between two continuous variables. Each point on the plot represents an observation, making it useful for identifying patterns, trends, and outliers.

2. **Bar Plot (`sns.barplot`):**
   - **Use:** Bar plots are useful for showing the distribution of a categorical variable. They display the average value or a summary statistic for different categories, making it easy to compare values between groups.

3. **Box Plot (`sns.boxplot`):**
   - **Use:** Box plots are effective for visualizing the distribution of a continuous variable across different categories. They display the median, quartiles, and potential outliers, providing insights into the central tendency and spread of the data.

4. **Heatmap (`sns.heatmap`):**
   - **Use:** Heatmaps are great for visualizing the relationship between two categorical variables. They use color intensity to represent the magnitude of a third variable, making it easy to identify patterns and correlations in a dataset.

5. **Pair Plot (`sns.pairplot`):**
   - **Use:** Pair plots are used to visualize pairwise relationships between multiple variables in a dataset. It plots scatter plots for numeric variables and histograms for univariate distributions along the diagonal. This is useful for identifying patterns and correlations in multivariate data.

These are just a few examples of the plots that Seaborn offers. Seaborn provides a variety of other plots and customization options to meet different visualization needs in data analysis.

## Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

# Add title and labels
plt.title('FMRI Dataset - Line Plot')
plt.xlabel('Timepoint')
plt.ylabel('Signal')

# Show the plot
plt.show()
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Create two box plots using Seaborn
plt.figure(figsize=(12, 6))

# Box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Box Plot of Age by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Age')

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Box Plot of Fare by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Fare')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

## Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram with hue for 'cut'
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', bins=30, multiple="stack", edgecolor='black')

# Add title and labels
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

## Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Create a pair plot with hue for 'species'
sns.set(style="ticks", color_codes=True)
sns.pairplot(iris_data, hue='species')

# Show the plot
plt.show()
```

## Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the dataset to create a matrix suitable for a heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)

# Add title and labels
plt.title('Flights Dataset - Heatmap')
plt.xlabel('Year')
plt.ylabel('Month')

# Show the plot
plt.show()
```
