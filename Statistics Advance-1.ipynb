## Q1. What is the Probability density function?


The Probability Density Function (PDF) is a concept in probability theory and statistics that describes the likelihood of a continuous random variable falling within a particular range of values. It is commonly used to characterize the probability distribution of a continuous random variable

It's important to note that for a continuous random variable, the probability of any specific value is technically zero. Instead, probabilities are associated with intervals, and the PDF provides a way to understand the likelihood of different intervals.

Common examples of probability density functions include the normal distribution, exponential distribution, and uniform distribution. The shape of the PDF can vary depending on the characteristics of the underlying random variable.

## Q2. What are the types of Probability distribution?

There are several types of probability distributions, and they can be broadly categorized into two main types: discrete probability distributions and continuous probability distributions.

1. **Discrete Probability Distributions:**
   - **Probability Mass Function (PMF):** Describes the probabilities of discrete outcomes in a discrete random variable.
   - **Examples:**
     - **Bernoulli Distribution:** Models a binary outcome (success/failure) with a probability \(p\) of success.
     - **Binomial Distribution:** Models the number of successes in a fixed number of independent Bernoulli trials.
     - **Poisson Distribution:** Models the number of events occurring in a fixed interval of time or space.

2. **Continuous Probability Distributions:**
   - **Probability Density Function (PDF):** Describes the probabilities of different outcomes in a continuous random variable.
   - **Examples:**
     - **Normal Distribution:** Symmetrical and bell-shaped. Describes many natural phenomena and is characterized by its mean and standard deviation.
     - **Exponential Distribution:** Models the time until an event occurs in a Poisson process.
     - **Uniform Distribution:** All values within a given range are equally likely.
     - **Gamma Distribution:** Generalizes the exponential distribution to model the time until \(k\) events occur in a Poisson process.

These are just a few examples, and there are many other probability distributions used in various fields such as statistics, physics, finance, and engineering. The choice of distribution depends on the characteristics of the data and the specific problem being addressed. Understanding probability distributions is fundamental in probability theory and statistics, as they provide a mathematical framework for modeling and analyzing uncertain or random phenomena.

## Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.

Certainly! You can use the scipy.stats.norm module to calculate the probability density function (PDF) of a normal distribution in Python. Here's a simple function that takes a value, mean, and standard deviation as input and returns the PDF:

```python
pip install scipy
```

```python
from scipy.stats import norm

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution.

    Parameters:
    - x: Value at which to evaluate the PDF.
    - mean: Mean of the normal distribution.
    - std_dev: Standard deviation of the normal distribution.

    Returns:
    - pdf: Probability density function value at the given point.
    """
    pdf = norm.pdf(x, loc=mean, scale=std_dev)
    return pdf

# Example usage:
mean_value = 0
std_dev_value = 1
point_value = 1.5

result = normal_pdf(point_value, mean_value, std_dev_value)
print(f"The PDF at {point_value} is: {result}")
```

## Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.

The Binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent and identically distributed Bernoulli trials. The key properties of a Binomial distribution are:

1. **Number of Trials (n):** The fixed number of identical trials is denoted by \(n\).
2. **Probability of Success (p):** The probability of success on each trial is denoted by \(p\).
3. **Probability of Failure (q):** The probability of failure on each trial, where \(q = 1 - p\).
4. **Independence:** The trials must be independent, meaning the outcome of one trial does not affect the outcome of another.
5. **Discreteness:** The random variable representing the number of successes is discrete.

The probability mass function (PMF) of the Binomial distribution is given by:

\[ P(X = k) = \binom{n}{k} \cdot p^k \cdot (1-p)^{n-k} \]

where \(X\) is the random variable representing the number of successes, \(k\) is the number of successes, \(\binom{n}{k}\) is the binomial coefficient, and \(p\) is the probability of success on a single trial.

**Examples of Events where Binomial Distribution can be Applied:**

1. **Coin Flipping:**
   - **Scenario:** Consider flipping a fair coin multiple times (e.g., 10 times).
   - **Binomial Variables:**
     - \(n\) (number of trials) = 10 (flips)
     - \(p\) (probability of heads) = 0.5
   - **Question:** What is the probability of getting exactly 7 heads?

2. **Quality Control:**
   - **Scenario:** In a manufacturing process, a product is either defective or non-defective.
   - **Binomial Variables:**
     - \(n\) (number of items inspected) = 100
     - \(p\) (probability of a defective item) = 0.03 (for example)
   - **Question:** What is the probability of finding less than 5 defective items in a sample of 100?

In both examples, the events involve a fixed number of independent trials with a binary outcome (success/failure), making them suitable for modeling with a Binomial distribution. The Binomial distribution is a versatile and widely used distribution in probability and statistics.

## Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(42)

# Parameters for the binomial distribution
n_trials = 1000  # Size of the random sample
p_success = 0.4  # Probability of success

# Generate a random sample from a binomial distribution
random_sample = np.random.binomial(n=1, p=p_success, size=n_trials)

# Plotting the histogram
plt.hist(random_sample, bins=[-0.5, 0.5, 1.5], edgecolor='black')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title(f'Binomial Distribution (n=1, p={p_success}) - Random Sample of {n_trials}')
plt.xticks([0, 1], ['Failure', 'Success'])
plt.show()
```

## Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.

```python
from scipy.stats import poisson

def poisson_cdf(x, mean):
    """
    Calculate the cumulative distribution function (CDF) of a Poisson distribution.

    Parameters:
    - x: Value at which to evaluate the CDF.
    - mean: Mean of the Poisson distribution.

    Returns:
    - cdf: Cumulative distribution function value at the given point.
    """
    cdf = poisson.cdf(x, mu=mean)
    return cdf

# Example usage:
mean_value = 3
point_value = 5

result = poisson_cdf(point_value, mean_value)
print(f"The CDF at {point_value} is: {result}")
```

## Q7. How Binomial distribution different from Poisson distribution?

The Binomial distribution and Poisson distribution are both probability distributions used to model the number of events, but they are applicable in different scenarios and have distinct characteristics. Here are the main differences between the Binomial and Poisson distributions:

1. **Nature of Events:**
   - **Binomial Distribution:** Models the number of successes in a fixed number of independent and identical Bernoulli trials. Each trial can result in a success or failure, and the trials are mutually exclusive.
   - **Poisson Distribution:** Models the number of events that occur in a fixed interval of time or space. The events are assumed to be rare and random, and they occur independently.

2. **Number of Trials:**
   - **Binomial Distribution:** Requires a fixed number of trials (\(n\)) to be specified in advance.
   - **Poisson Distribution:** Does not involve a fixed number of trials; it is defined over a fixed interval of time or space.

3. **Probability of Success:**
   - **Binomial Distribution:** Involves a probability of success (\(p\)) on each trial.
   - **Poisson Distribution:** Does not directly involve a probability of success. Instead, it is characterized by the average rate of events (\(\lambda\)).

4. **Applicability:**
   - **Binomial Distribution:** Applicable when there is a fixed number of trials, each with a binary outcome, and the probability of success is constant across trials. Commonly used in scenarios such as coin flipping or quality control.
   - **Poisson Distribution:** Applicable when events occur randomly and independently over time or space, and the events are rare. Commonly used in scenarios such as the number of phone calls received at a call center in an hour.

5. **Shape of Distribution:**
   - **Binomial Distribution:** Can be skewed or symmetric, depending on the values of \(n\) and \(p\). As \(n\) increases and \(p\) approaches 0.5, the Binomial distribution becomes more symmetric.
   - **Poisson Distribution:** Often right-skewed, especially when the average rate (\(\lambda\)) is low.

6. **Limiting Behavior:**
   - **Binomial Distribution:** As the number of trials (\(n\)) becomes large, and the probability of success (\(p\)) becomes small in a way that \(np\) remains constant, the Binomial distribution approaches the Poisson distribution.
   - **Poisson Distribution:** Often arises as an approximation to the Binomial distribution when \(n\) is large and \(p\) is small.

In summary, while both distributions are used to model the number of events, the key differences lie in the nature of the events, the number of trials, and the underlying assumptions about the process generating the events. The Binomial distribution is suitable for a fixed number of trials with binary outcomes, while the Poisson distribution is suitable for modeling rare, random events over a fixed interval.

## Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.

```python
import numpy as np

# Set the seed for reproducibility
np.random.seed(42)

# Parameters for the Poisson distribution
mean_value = 5  # Mean of the distribution
sample_size = 1000  # Size of the random sample

# Generate a random sample from a Poisson distribution
poisson_sample = np.random.poisson(lam=mean_value, size=sample_size)

# Calculate sample mean and variance
sample_mean = np.mean(poisson_sample)
sample_variance = np.var(poisson_sample, ddof=1)  # ddof=1 for sample variance

# Print results
print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")
```

##  Q9. How mean and variance are related in Binomial distribution and Poisson distribution?

In both the Binomial distribution and Poisson distribution, the mean (\(\mu\)) and variance (\(\sigma^2\)) are closely related, and their relationship is influenced by the characteristics of the specific distribution.

**Binomial Distribution:**

For a Binomial distribution with parameters \(n\) (number of trials) and \(p\) (probability of success on each trial):

1. **Mean (\(\mu\)):** The mean of a Binomial distribution is given by \(\mu = np\), where \(n\) is the number of trials and \(p\) is the probability of success on each trial.

2. **Variance (\(\sigma^2\)):** The variance of a Binomial distribution is given by \(\sigma^2 = np(1 - p)\).

3. **Relationship:** The relationship between the mean and variance in a Binomial distribution is \(\sigma^2 = np(1 - p)\). The variance increases as either the number of trials (\(n\)) or the probability of success (\(p\)) deviates from 0.5.

**Poisson Distribution:**

For a Poisson distribution with parameter \(\lambda\) (average rate of events per unit interval):

1. **Mean (\(\mu\)):** The mean of a Poisson distribution is given by \(\mu = \lambda\).

2. **Variance (\(\sigma^2\)):** The variance of a Poisson distribution is also given by \(\sigma^2 = \lambda\).

3. **Relationship:** In a Poisson distribution, the mean and variance are equal (\(\mu = \sigma^2 = \lambda\)). This characteristic is specific to the Poisson distribution and is a result of the underlying assumptions about the rarity and randomness of events.

**Summary:**

- In both the Binomial and Poisson distributions, the mean and variance are measures of central tendency and dispersion, respectively.
  
- For the Binomial distribution, the variance is influenced by both the number of trials (\(n\)) and the probability of success (\(p\)).

- For the Poisson distribution, the variance is solely determined by the average rate of events (\(\lambda\)), and the mean and variance are equal.

- In the Poisson distribution, the equality of mean and variance (\(\mu = \sigma^2\)) is a distinctive feature, reflecting the assumption that events occur independently and randomly at a constant average rate.

## Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

In a normal distribution, also known as a Gaussian distribution or bell curve, the data is symmetrically distributed around the mean. The mean is the central point of the distribution, and the data is more densely packed around the mean. The shape of the normal distribution is such that as you move away from the mean in either direction, the frequency of data points decreases.

Specifically:

- The least frequent data appears in the tails of the distribution, farthest away from the mean.
- The tails of the normal distribution extend infinitely in both directions.

In a standard normal distribution (with a mean of 0 and a standard deviation of 1), the data becomes increasingly rare as you move farther from the mean. The probability density is higher around the mean and decreases as you move towards the tails.

In summary, the least frequent data points in a normal distribution are found in the extreme tails, both on the left (negative side) and right (positive side) of the mean.
